## Alter
- Drop rows.
- Drop columns.
- Slice out data based on some conditions.
- Alter values based on some criteria.

```python
# Dropping missing values.
df.isna().sum()
```

```python
no_missing_age = df.dropna(subset=['age'])
no_missing_age.isna().sum()
```

```python
df.shape, df.dropna(thresh=14).shape
```

```python
df.dropna(subset=['age'], inplace=True)
df.isna().sum()
```

```python
# Dropping rows.
df.drop(list(range(0, 10, 2)))
```

```python
# Dropping columns.
df.drop('embark_town', axis=1)
```

```python
df
```

```python
# Slicing data based on a condition. (using square brackets)
df[((df['survived'] == 1) & (df['age'] > 10))]
```

```python
# Preferred method (using .loc)
df.loc[:, ['survived', 'fare']]
```
